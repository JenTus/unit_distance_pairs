```python
import itertools
```

```python
five = list(itertools.combinations([i for i in range(7)], 5))
[list(itertools.combinations(two, 2)) for two in five]
```

```python
four = list(itertools.combinations([i for i in range(8)], 4))
[list(itertools.combinations(two, 2)) for two in four]
```

```python
all_possible_list = [i for i in itertools.product([0,1], repeat=21) if sum(i) >= 10]
```
